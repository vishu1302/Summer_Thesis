```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***Basic Graph***

```python
x=list(np.linspace(1,11,5))
y=list(np.linspace(2,21,5))


#keyword argument notation
plt.plot(x,y, label ='line', color='r', linewidth=3, marker='.',markersize=10)


#linestyle='--' #markeredgecolor='blue'
plt.title('Our First Graph', fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#X and Y Tickmarks (scale on your graph)
#plt.xticks([0,1,2]) #for making customized ticks
#plt.yticks([0,2,4,6,7.5])

plt.legend()
plt.show()
```

```python
x1=list(np.linspace(1,11,5))
y=list(np.linspace(2,21,5))

#use short hand notatations
#fmt ='[color][marker][line]'

plt.plot(x1,y, 'r.--' ,label ='2x')
plt.legend()
plt.show()
```

```python
x=list(np.linspace(1,11,5))
y=list(np.linspace(2,21,5))

#use short hand notatations
#fmt ='[color][marker][line]'



plt.plot(x,y, 'r.--' ,label ='line')

print(y)

#line number two
x2= np.arange(0,10,2)
plt.plot(x2,x2**2, 'b^-',label='xsquare')

plt.title('Second Graph')
plt.legend()
plt.show()
```

```python
#projection

x=list(np.linspace(1,11,5))
y=list(np.linspace(2,21,5))

#use short hand notatations
#fmt ='[color][marker][line]'


#Resize your Graph
#dpi specifies pixels per inch (While saving use 300)

plt.figure(figsize=(5,3), dpi=100)

plt.plot(x,y, 'r.--' ,label ='line')

print(y)

#line number two
x2= np.arange(0,10,1)
#plotting the main  line
plt.plot(x2[:7],x2[:7]**2, 'b^-',label='xsquare')

#plotting the projection line
plt.plot(x2[6:],x2[6:]**2, 'b^--')

plt.title('Second Graph')
plt.legend()

plt.savefig('mygraph.png',dpi=300)
plt.show()
```

bar chart

```python
labels =['A','B','C']
values =[1,5,3]
bars=plt.bar(labels,values)

patterns=['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#ars[0].set_hatch('/')

plt.figure(figsize=(6,4))
plt.show()
```
